```python
import pandas as pd

df = pd.read_excel('Tableau Full Project.xlsx')

try:
    file_path = 'Tableau Full Project.xlsx'
    dfs = pd.read_excel(file_path, sheet_name = None)
    print("Sheets loaded:", dfs.keys())

    listings_df = dfs["Listings"]
    reviews_df = dfs["Reviews"]
    calendar_df = dfs["Calendar"]

    print("\nListings Sheet:")
    display(listings_df.head())

    print("\nReviews Sheet:")
    display(reviews_df.head())

    print("\nCalendar Sheet:")
    display(calendar_df.head())

except FileNotFoundError:
    print("Error: 'Tableau Full Project.xlsx' not found.")
    dfs = None
except Exception as e:
    print(f"An error occurred: {e}")
    dfs = None
```
